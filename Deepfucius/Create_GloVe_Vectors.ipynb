```python
import itertools
from glove import Corpus, Glove
import csv
import nltk
from gensim.models import word2vec
```

```python
sentences2 = []
paragraph = open('Outputs/out_file1.csv')

with open('Outputs/out_file1.csv') as csvfile:
    reader = csv.reader(csvfile)
    sentences2 = [item for sublist in reader for item in sublist]
```

```python
sentences = [nltk.word_tokenize(sent) for sent in sentences2]
print(sentences2)
```

```python
test = word2vec.Text8Corpus('text8')
```

```python
corpus = Corpus()
corpus.fit(sentences, window = 10)
```

```python

glove = Glove(no_components = 100, learning_rate = 0.05)

glove.fit(corpus.matrix, epochs = 30, no_threads = 4, verbose = True)
```

```python
glove.add_dictionary(corpus.dictionary)

glove.most_similar('they')
```

```python
print(glove.dictionary)
```

```python
Vectors = {word: glove.word_vectors[glove.dictionary[word]] for word in glove.dictionary.keys()}
```
